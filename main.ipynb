```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso
```

```python
df=pd.read_csv("car data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        print(df[col].value_counts())
```

```python
df.replace({"Fuel_Type": {"Diesel": 0, "Petrol": 1, "CNG": 2}}, inplace=True)
df.replace({"Seller_Type": {"Dealer": 0, "Individual": 1}}, inplace=True)
df.replace({"Transmission": {"Manual": 0, "Automatic": 1}}, inplace=True)
```

```python
df.head()
```

```python
X = df.drop(['Car_Name','Selling_Price'],axis=1)
Y= df["Selling_Price"]
```

```python
X.info()
```

```python
Y.info()
```

```python
lin_reg=LinearRegression()
lasso_reg=Lasso() 
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
lin_reg.fit(x_train, y_train)
```

```python
train_pred = lin_reg.predict(x_train)
e_score_train = metrics.r2_score(y_train, train_pred)
print(f"Training R^2 Score: {e_score_train}")
```

```python
test_pred = lin_reg.predict(x_test)
e_score_test = metrics.r2_score(y_test, test_pred)
print(f"Testing R^2 Score: {e_score_test}")
```

```python
lasso_reg.fit(x_train, y_train)
```

```python
train_pred_lasso = lasso_reg.predict(x_train)
e_score_train_lasso = metrics.r2_score(y_train, train_pred_lasso)
print(f"Lasso Training R^2 Score: {e_score_train_lasso}")
```

```python
test_pred_lasso = lasso_reg.predict(x_test)
e_score_test_lasso = metrics.r2_score(y_test, test_pred_lasso)
print(f"Lasso Testing R^2 Score: {e_score_test_lasso}")
```
